```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from google.colab import drive

drive.mount('/content/drive')
```

```python
#Update path as necessary
df = pd.read_csv("/content/drive/MyDrive/DS4B/Ames/train.csv")
df.head()
```

```python
df.describe()
```

```python
# Technique 1: Histogram for SalePrice
plt.figure(figsize=(8, 6))
sns.histplot(df['SalePrice'], kde=True, color='blue', bins=30)
plt.title('Histogram of SalePrice')
plt.xlabel('Sale Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Technique 3: Boxplot for OverallQual vs SalePrice
plt.figure(figsize=(10, 6))
sns.boxplot(x='OverallQual', y='SalePrice', data=df)
plt.title('Boxplot of SalePrice by Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Technique 4: Scatterplot for GrLivArea vs SalePrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x='GrLivArea', y='SalePrice', data=df)
sns.regplot(x='GrLivArea', y='SalePrice', data=df, scatter=False, color='red')  # Add regression line without scatter points
plt.title('Scatterplot of SalePrice vs Above Grade Living Area with Linear Regression Line')
plt.xlabel('Above Grade Living Area (sqft)')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Technique 5: Barplot for Neighborhood vs SalePrice
plt.figure(figsize=(12, 8))
sns.barplot(x='Neighborhood', y='SalePrice', data=df)
plt.title('Average Sale Price by Neighborhood')
plt.xlabel('Neighborhood')
plt.ylabel('Average Sale Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Technique 6: Pairplot for selected numeric features
sns.pairplot(df[['SalePrice', 'OverallQual', 'GrLivArea', 'TotalBsmtSF', 'YearBuilt']])
plt.show()
```

```python
# Technique 7: Violin plot for OverallCond vs SalePrice
plt.figure(figsize=(10, 6))
sns.violinplot(x='OverallCond', y='SalePrice', data=df)
plt.title('Violinplot of SalePrice by Overall Condition')
plt.xlabel('Overall Condition')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Technique 8: Countplot for HouseStyle
plt.figure(figsize=(8, 6))
sns.countplot(x='HouseStyle', data=df)
plt.title('Count of House Styles')
plt.xlabel('House Style')
plt.ylabel('Count')
plt.xticks(rotation=45)
plt.show()
```

```python
# Technique 9: Boxplot for YearBuilt vs SalePrice
plt.figure(figsize=(16, 8))
sns.boxplot(x='YearBuilt', y='SalePrice', data=df)
plt.title('Boxplot of SalePrice by Year Built')
plt.xlabel('Year Built')
plt.ylabel('Sale Price')
plt.xticks(rotation=90)
plt.show()
```

```python
# Technique 10: Swarmplot for MSZoning vs SalePrice
plt.figure(figsize=(10, 6))
sns.swarmplot(x='MSZoning', y='SalePrice', data=df)
plt.title('Swarmplot of SalePrice by Zoning Classification')
plt.xlabel('Zoning Classification')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Technique 11: Boxplot for OverallQual vs GrLivArea
plt.figure(figsize=(10, 6))
sns.boxplot(x='OverallQual', y='GrLivArea', data=df)
plt.title('Boxplot of Above Grade Living Area by Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Above Grade Living Area (sqft)')
plt.show()
```

```python
# Technique 12: Barplot for MSZoning vs SalePrice
plt.figure(figsize=(10, 6))
sns.barplot(x='MSZoning', y='SalePrice', data=df)
plt.title('Average Sale Price by Zoning Classification')
plt.xlabel('Zoning Classification')
plt.ylabel('Average Sale Price')
plt.show()
```

```python
# Technique 13: Scatterplot for LotArea vs SalePrice
plt.figure(figsize=(10, 6))
sns.scatterplot(x='LotArea', y='SalePrice', data=df)
sns.regplot(x='LotArea', y='SalePrice', data=df, scatter=False, color='red')  # Add regression line without scatter points
plt.title('Scatterplot of SalePrice vs Lot Area with Linear Regression Line')
plt.xlabel('Lot Area (sqft)')
plt.ylabel('Sale Price')
plt.show()
```

```python
# Technique 14: Countplot for BldgType
plt.figure(figsize=(8, 6))
sns.countplot(x='BldgType', data=df)
plt.title('Count of Building Types')
plt.xlabel('Building Type')
plt.ylabel('Count')
plt.show()
```

```python
# Technique 15: Violinplot for OverallQual vs GrLivArea
plt.figure(figsize=(10, 6))
sns.violinplot(x='OverallQual', y='GrLivArea', data=df)
plt.title('Violinplot of Above Grade Living Area by Overall Quality')
plt.xlabel('Overall Quality')
plt.ylabel('Above Grade Living Area (sqft)')
plt.show()
```

```python
# Technique 16: Boxplot for LotShape vs SalePrice
plt.figure(figsize=(10, 6))
sns.boxplot(x='LotShape', y='SalePrice', data=df)
plt.title('Boxplot of SalePrice by Lot Shape')
plt.xlabel('Lot Shape')
plt.ylabel('Sale Price')
plt.show()
```
